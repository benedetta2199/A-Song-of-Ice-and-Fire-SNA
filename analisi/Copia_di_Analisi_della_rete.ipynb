# **La rete è connessa o frammentata?**
La connettività della rete è un indicatore chiave per comprendere la coesione complessiva del gruppo. Una rete fortemente connessa suggerisce un ambiente in cui i personaggi interagiscono frequentemente e direttamente, mentre una rete frammentata implica sottogruppi isolati o fazioni separate.

*   **Densità**: Misura la proporzione delle connessioni effettive rispetto al numero massimo possibile di connessioni nella rete. Una densità alta indica una rete altamente interconnessa.

*   **Clusterizzazione globale**: Indica la tendenza dei nodi a formare gruppi o sottogruppi interconnessi. Un alto coefficiente di clusterizzazione suggerisce che la maggior parte dei personaggi all'interno dei sottogruppi è direttamente collegato agli altri personaggi del suo sottogruppo.

Questa parte la userei per dire lo scale coso, il centro connesso e il resto sparso

```python
import networkx as nx
import matplotlib.pyplot as plt

def analyze_and_plot_graph(G, threshold):
    # Calcolare il degree di ciascun nodo
    node_degrees = dict(G.degree())

    # Filtrare i nodi che soddisfano la soglia
    filtered_nodes = [node for node, degree in node_degrees.items() if degree >= threshold]

    # Creare un sottografo contenente solo i nodi filtrati
    filtered_graph = G.subgraph(filtered_nodes)

    # Ricalcolare la densità e il coefficiente di clusterizzazione per la rete filtrata
    filtered_density = nx.density(filtered_graph)
    filtered_global_clustering = nx.average_clustering(filtered_graph)

    # --- PLOTTA L'INTERO GRAFO ---
    # Impostare il layout per l'intero grafo
    pos = nx.spring_layout(G, seed=42)

    # Creare la visualizzazione dell'intero grafo
    plt.figure(figsize=(16, 8))

    plt.subplot(121)
    # Dimensione dei nodi basata sul degree, scalata tra 5 e 200
    node_sizes = [2 + (node_degrees[node] / max(node_degrees.values())) * 100 for node in G.nodes()]

    # Disegnare i nodi che non sono stati filtrati
    other_nodes = [node for node in G.nodes() if node not in filtered_nodes]
    nx.draw_networkx_nodes(G, pos, nodelist=other_nodes, node_size=[2 + (node_degrees[node] / max(node_degrees.values())) * 100 for node in other_nodes], node_color='slategray', alpha=0.6)

    # Disegnare i nodi filtrati
    nx.draw_networkx_nodes(G, pos, nodelist=filtered_nodes, node_size=[2 + (node_degrees[node] / max(node_degrees.values())) * 100 for node in filtered_nodes], node_color='navy', alpha=0.8)

    # Disegnare gli archi
    nx.draw_networkx_edges(G, pos, alpha=0.3, width=0.4, edge_color='gray')

    plt.title("Entire Graph with Highlighted Filtered Nodes", fontsize=15)
    plt.axis('off')

    # --- PLOTTA IL GRAFO FILTRATO ---
    plt.subplot(122)
    # Impostare il layout per il sottografo
    pos_filtered = nx.spring_layout(filtered_graph, seed=42)

    # Disegnare gli archi
    nx.draw_networkx_edges(filtered_graph, pos_filtered, alpha=0.3, width=0.4, edge_color='gray')
    nx.draw_networkx_nodes(filtered_graph, pos_filtered, node_size=[5 + (node_degrees[node] / max(node_degrees.values())) * 195 for node in filtered_nodes], node_color='navy', alpha=0.8)

    # Aggiungere il titolo
    plt.title(f"Filtered Network", fontsize=15)

    plt.suptitle(f"Density: {filtered_density:.3f} - Global Clustering: {filtered_global_clustering:.3f} - Threshold {threshold}%", fontsize=25)
    plt.axis('off')
    plt.show()
```

# **Quanto è efficiente la rete?**
L'efficienza della rete determina quanto rapidamente le informazioni o influenze possono propagarsi da un personaggio all'altro. Una rete efficiente facilita lo scambio di informazioni tra i personaggi con un minor numero di intermediari.

*   **Diametro**: Rappresenta la distanza massima tra due nodi nella rete. Un diametro minore indica una maggiore efficienza di trasmissione.
*   **Compattezza (distanza media)**: Rappresenta la distanza media tra tutti i nodi della rete. Valori bassi indicano una rete compatta, con brevi percorsi di comunicazione tra i personaggi.


```python
# 2) Problema: Quanto è efficiente la rete nel trasmettere informazioni?
# Metriche usate: Diametro, compactness (distanza media).

#poichè sappiamo che la rete è conessa possiamo direttamente calcolare il diametro
diameter = nx.diameter(G)
print(f"Diametro della rete: {diameter}")

# Calcolare la distanza media (compactness)
avg_shortest_path = nx.average_shortest_path_length(G)
print(f"Distanza media della rete: {avg_shortest_path}")


# Istogramma delle distanze più brevi tra i nodi
# Otteniamo tutte le distanze più brevi tra le coppie di nodi
all_shortest_paths = dict(nx.all_pairs_shortest_path_length(G))
distances = []

# Aggiungiamo le distanze in una lista
for source, targets in all_shortest_paths.items():
    for target, dist in targets.items():
        if source != target:  # Evitiamo di includere la distanza a se stesso
            distances.append(dist)

# Plot dell'istogramma delle distanze con Seaborn
plt.figure(figsize=(6, 4))  # Increased figure size for better visibility
sns.histplot(distances, bins=range(1, diameter + 2), edgecolor='black', linewidth=0.2, zorder=2)
# Added color, edgecolor, and linewidth for aesthetics
plt.title(f"Distribution of Shortest Distances with Network Diameter {diameter}", fontsize=12)
# Increased title fontsize and added fontweight
plt.xlabel('Distance between Nodes', fontsize=10)  # Increased xlabel fontsize
plt.ylabel('Frequency', fontsize=10)  # Increased ylabel fontsize
plt.xticks(fontsize=7)  # Increased xticks fontsize
plt.yticks(fontsize=7)  # Increased yticks fontsize
plt.grid(axis='y', alpha=0.5, zorder=1) # Added a subtle grid for better readability
plt.tight_layout()  # Adjusts subplot params for a tight layout
plt.show()
```

# **Quali nodi o collegamenti sono critici per la struttura della rete?**
Individuare i nodi o le connessioni critiche permette di comprendere quali personaggi o relazioni siano fondamentali per mantenere l'integrità della rete. La rimozione di questi elementi potrebbe frammentare la rete o ridurre significativamente la sua efficienza.

*  **Connettività dei nodi**: Valuta l'importanza di ogni nodo per la coesione della rete. Personaggi con alta connettività sono punti focali critici.
*  **Connettività degli archi**: Misura l'importanza delle connessioni tra i nodi. Le relazioni critiche sono quelle la cui rimozione avrebbe un forte impatto sulla struttura della rete.



```python
# Trovare i nodi la cui rimozione disconnette la rete
articulation_points = list(nx.articulation_points(G))

print(f"Nodi critici (articulation points): {articulation_points}")
print(f"Numero di nodi critici: {len(articulation_points)}")

# Rimuovere un articulation point per verificare l'effetto sulla connettività
test_graph = G.copy()
test_graph.remove_node(articulation_points[0])  # Rimuoviamo il primo nodo critico

# Verificare il numero di componenti connesse dopo la rimozione
num_components_after_removal = nx.number_connected_components(test_graph)
print(f"Numero di componenti connesse dopo la rimozione di {articulation_points[0]}: {num_components_after_removal}")
```

```python
# Trovare gli archi critici (bridge edges)
bridge_edges = list(nx.bridges(G))

print(f"Archi critici (bridge edges): {bridge_edges}")
print(f"Numero di archi critici: {len(bridge_edges)}")

# Rimuovere un arco critico per verificare l'effetto sulla connettività
test_graph = G.copy()
test_graph.remove_edge(*bridge_edges[0])  # Rimuoviamo il primo arco critico

# Verificare il numero di componenti connesse dopo la rimozione
num_components_after_removal = nx.number_connected_components(test_graph)
print(f"Numero di componenti connesse dopo la rimozione dell'arco {bridge_edges[0]}: {num_components_after_removal}")
```

```python
# Ordinare i nodi per centralità di prossimità in ordine decrescente
sorted_nodes = sorted(articulation_points, key=lambda node: G.nodes[node]['closnesscentrality'], reverse=True)
# Ottenere i primi n nodi con la maggiore centralità di prossimità
top_n_nodes = sorted_nodes[:10]

# Impostare il layout del grafo
pos = nx.spring_layout(G, seed=42)

# Evidenziare i nodi critici
node_colors = ['red' if node in articulation_points else 'silver' for node in G.nodes()]

# Evidenziare gli archi critici
edge_colors = ['red' if edge in bridge_edges else 'silver' for edge in G.edges()]
edge_alpha = [0.6 if edge in bridge_edges else 0.4 for edge in G.edges()]

# Creare la visualizzazione
plt.figure(figsize=(16, 9))

# Disegnare gli archi
nx.draw_networkx_nodes(G, pos=pos, node_size=7, node_color=node_colors, alpha=0.8)
nx.draw_networkx_edges(G, pos=pos, alpha=edge_alpha, width=0.4, edge_color=edge_colors)

# Aggiungere le etichette per i primi n nodi con la maggiore centralità di prossimità
labels = {node: node for node in top_n_nodes}

# Disegnare le etichette con bbox (sfondo) e font ridotto
texts = []
for node, label in labels.items():
    text = plt.text(pos[node][0], pos[node][1], label, fontsize=7, color="maroon",
                    ha='center', va='center', bbox=dict(facecolor='white', alpha=0.75, edgecolor='none', boxstyle='round,pad=0.2'))
    texts.append(text)

# Utilizzare adjust_text per regolare automaticamente le etichette
adjust_text(texts, force_text=0.2, expand_text=(1.05, 1.05), only_move={'points': 'y', 'texts': 'xy'})

plt.title("Critical Nodes and Edges in the Network", fontsize=25)
plt.axis('off')
plt.show()
```

```python
from collections import Counter

# Dizionario per memorizzare le combinazioni di componenti e la loro frequenza
component_combinations = Counter()

for node in articulation_points:
    # Copiare il grafo e rimuovere il nodo critico
    test_graph = G.copy()
    test_graph.remove_node(node)

    # Ottenere le componenti connesse dopo la rimozione
    components = list(nx.connected_components(test_graph))

    # Calcolare le dimensioni delle componenti e ordinarle in modo decrescente
    component_sizes = sorted([len(component) for component in components], reverse=True)

    # Aggiungere la combinazione al conteggio
    component_combinations[tuple(component_sizes)] += 1

# Creare un istogramma delle combinazioni più frequenti
plt.figure(figsize=(10, 6))
combinations, counts = zip(*component_combinations.items())  # Separare chiavi e valori
labels = [str(comb) for comb in combinations]  # Convertire le combinazioni in stringhe per l'asse x

plt.bar(labels, counts, color='skyblue', edgecolor='black')
plt.xticks(rotation=90, fontsize=8)
plt.title("Distribution of Component Size Combinations", fontsize=15)
plt.xlabel("Size Combinations", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

SMALL WORLDNESS

```python
import networkx as nx
import numpy as np

# Supponiamo che G sia il grafo

# Calcola il coefficiente di clustering medio e la lunghezza media del cammino nella rete reale
C_real = nx.average_clustering(G)
try:
    L_real = nx.average_shortest_path_length(G)
except nx.NetworkXError:
    print("Il grafo non è connesso. Calcoliamo la lunghezza media del cammino per la componente più grande.")
    largest_cc = max(nx.connected_components(G), key=len)
    G_largest = G.subgraph(largest_cc)
    L_real = nx.average_shortest_path_length(G_largest)

# Genera un grafo casuale con lo stesso numero di nodi e gradi (grafo di configurazione)
G_random = nx.configuration_model([d for n, d in G.degree()])

# Rimuove multiarchi e self-loop dal grafo casuale
G_random = nx.Graph(G_random)  # Converte in un grafo semplice
G_random.remove_edges_from(nx.selfloop_edges(G_random))

# Calcola il coefficiente di clustering medio e la lunghezza media del cammino nel grafo casuale
C_random = nx.average_clustering(G_random)
try:
    L_random = nx.average_shortest_path_length(G_random)
except nx.NetworkXError:
    print("Il grafo casuale non è connesso. Calcoliamo la lunghezza media del cammino per la componente più grande.")
    largest_cc_random = max(nx.connected_components(G_random), key=len)
    G_random_largest = G_random.subgraph(largest_cc_random)
    L_random = nx.average_shortest_path_length(G_random_largest)

# Calcola il valore di small-worldness
small_worldness = (C_real / C_random) / (L_real / L_random)

# Risultati
print("Small-Worldness:")
print(f"C_real: {C_real}, L_real: {L_real}")
print(f"C_random: {C_random}, L_random: {L_random}")
print(f"Small-Worldness (S): {small_worldness}")
```
